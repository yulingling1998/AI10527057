```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt


delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z = X*X+Y*Y
```

```python
fig, ax = plt.subplots()

CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title('Simplest default with labels')
plt.hold(True)

V = np.array([[2,0],[3,0],[4,1],[5,1],[0,1],[7,1]])
origin = [-2], [-2] # origin point

plt.quiver(*origin, V[1,0], V[1,0], color=['r','r','r'], scale=21)
plt.show()
```
